### 1. Import Libraries

```python
import tensorflow as tf
from tensorflow import keras
```

### 2. Load the Dataset

```python
mnist= tf.keras.datasets.fashion_mnist
(traning_images,traning_labels),(test_images,test_labels)=mnist.load_data()
```

### 3. Define Class Labels

```python
class_names = [
    "T-shirt", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]
```

##  Neural Network Structure (Fashion MNIST Classifier)

###  Step 1: Normalize the Data

```python
traning_images=traning_images/255.0
test_images= test_images/255.0 
```

### Step 2: Visualize Data

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(class_names[traning_labels[i]])
    # plt.grid(False)
    # plt.imshow(traning_images[i])
    plt.imshow(traning_images[i],cmap=plt.cm.binary)
```

### Step 3: Define the Neural Network Model

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),           # Converts 2D 28x28 to 1D 784
    keras.layers.Dense(128, activation='relu'),           # Fully connected layer with 128 neurons
    keras.layers.Dense(10, activation='softmax')          # Output layer: 10 classes (probabilities)
])
```

### Step 4: Compile the Model & Train it

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
              )
model.fit(traning_images,traning_labels,epochs=10)
```

### Step 6: Evaluate Model Performance

```python
test_loss,test_acc= model.evaluate(test_images,test_labels, verbose=2)
print("Test accuracy" , test_acc)
```

### Step 7: Make Predictions

```python
import numpy as np
predict= model.predict(test_images)
np.argmax(predict[6])
```
